# EDA para o dataset Olist

Baseado no dataset da Olist foi pedido para fazer um modelo para prever quais usuários têm maior probabilidade de fazer uma segunda compra em até 30 dias após a realização da primeira. No entando para este exercício o problema requer apenas que seja criado um dataset básico que servirá de esqueleto para um possível modelo futuro.

Para este problemas precisamos análisar o conjunto de datasets disponíveis e procurar clientes que fizeram mais de uma compra, sendo que a segunda tenha sido em um intervalo menor que 30 dias.

Para este problema usaremos o polars, apenas por critério de aprendizado.

```python
#importando o pacote para o ambiente
import polars as pl
```

```python
# importando os arquivos .csv que serão usados na análise, depois de uma análise prévia dos dados presentes nos datasets disponibilizados

df_custumers = pl.read_csv(file = "/content/drive/MyDrive/Colab_Notebooks/Olist/Datasets/raw_datasets/olist_customers_dataset.csv")
df_order_dataset = pl.read_csv(file = '/content/drive/MyDrive/Colab_Notebooks/Olist/Datasets/raw_datasets/olist_orders_dataset.csv', parse_dates=True)
```

```python
# conhecendo um dos datasets
print(df_order_dataset.shape)
df_order_dataset.head()
```

```python
# Conhecendo o segundo dataset e verificando seu formato
print(df_custumers.shape)
df_custumers.head()
```

```python
#unindo as tabelas com datas de compras e id unicos dos clientes, usando um join left
df_unique_id_date = df_order_dataset[['order_id', 'customer_id', 'order_purchase_timestamp']].join(df_custumers[['customer_id', 'customer_unique_id']], on="customer_id", how="left")
df_unique_id_date.head()
```

```python
df_unique_id_date.shape
```

```python
#agrupando as compras dos clientes que compraram mais de uma vez
df = df_unique_id_date.groupby('customer_unique_id').agg(pl.col('order_purchase_timestamp'))
df.shape
```

```python
df.head()
```

```python
# duration (difference) between 2 dates (in microseconds)
col = pl.col("order_purchase_timestamp")

diff = pl.duration(microseconds=(col.arr.get(0) - col.arr.get(1)))
```

```python
# Filtrando somente clientes com mais de uma compra e criando a coluna com a feature 'target'
# Para executar alguma ação na coluna com base na condição, pode-se usar pl.when -> then -> caso contrário construa.
df_final = df.filter(col.arr.lengths() >= 2).with_columns(
    col.arr.sort(reverse=True).arr.slice(0, 2)
).with_columns(
    target = pl.when(diff <= pl.duration(days=30))\
                .then(1).otherwise(0)
)
```
